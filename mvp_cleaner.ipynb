```python
import pandas as pd
import csv
```

```python
#loading the dataset
df = pd.read_csv('C:/Users/colek/baseball_awards.csv')
df.head()
```

```python
df['Lg'] = None 
df.insert(0, 'Lg', df.pop('Lg')) #move Lg to leftmost side

df['Pos_Int'] = None 
df.insert(0, 'Pos_Int', df.pop('Pos_Int')) #move Lg to leftmost side

df['Pos'] = None 
df.insert(0, 'Pos', df.pop('Pos')) #move Lg to leftmost side

df['LgYear'] = None 
df.insert(0, 'LgYear', df.pop('LgYear')) #move Lg to leftmost side

df['IsMvp_Int'] = None 
df.insert(0, 'IsMvp_Int', df.pop('IsMvp_Int')) #move Rank to leftmost side

df['IsMvp'] = None 
df.insert(0, 'IsMvp', df.pop('IsMvp')) #move Rank to leftmost side

df['Rank'] = None 
df.insert(0, 'Rank', df.pop('Rank')) #move Rank to leftmost side

df.head()
```

```python
rnk = 0
lg = "NL"

for row in df.itertuples(index=True):
    year = row.Year
    name = row.Name
    index = row.Index
    era = row.ERA
    
    if pd.isna(name):
        rnk = 0
    df.at[index, 'Rank'] = rnk
    
    if (rnk == 1):
        df.at[index, 'IsMvp'] = "Yes"
        df.at[index, 'IsMvp_Int'] = 1
    else:
        df.at[index, 'IsMvp'] = "No"
        df.at[index, 'IsMvp_Int'] = 0

    if pd.isna(era):
        df.at[index, 'Pos'] = "Batter"
        df.at[index, 'Pos_Int'] = 1
    else:
        df.at[index, 'Pos'] = "Pitcher"
        df.at[index, 'Pos_Int'] = 0

    if pd.isna(name):
        if (lg == "NL"):
            lg = "AL"
        else:
            lg = "NL"
    df.at[index, 'Lg'] = lg
    df.at[index, 'LgYear'] = lg + str(year)[2:]

    rnk += 1
```

```python
df.dropna(subset=['Name'], inplace=True)
df.head(30)
```

```python
df.to_csv('baseball_mvp_cleaned.csv', index=False)
```
